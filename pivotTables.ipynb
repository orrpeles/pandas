## Working with seaborn directory to understand pivot tables

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(10)
```

```python
titanic.head
```

### These next 2 examples achieve the same result, 1.) using groupby 2.) using pivot_table

```python
titanic.groupby(['sex', 'class', 'embark_town'])[['survived']].aggregate('mean').unstack()
```

```python
titanic.pivot_table('survived', index = 'sex', columns = 'class')
```

### summary of all columns ('margins = True')

```python
titanic.pivot_table('survived', index = 'sex', columns = 'class', margins = True)
```

### looking at multilevel index dimension (age)

```python
age = pd.cut(titanic['age'], [0,18,80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

### looking at multivelel columns dimention (fare)

```python
fare = pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived', ['sex', age], ['class', fare])
```

```python
titanic.pivot_table(index = 'sex', columns = 'class', aggfunc = {'survived': sum, 'fare': 'mean'})
```
